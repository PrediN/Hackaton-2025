```python
import pandas as pd

csv = pd.read_csv("frequencia.csv")
df = pd.DataFrame(csv)

print(df.head(5))
i = 1
```

```python
df['bool_entrada'] = df['Entrada'].isnull()
df['bool_saida'] = df['Saida'].isnull()

for row, linha in df.iterrows():
    match (linha['bool_entrada'], linha['bool_saida']):
        case (True, True):
            print('Falta')
            df.at[row, 'Falta_analisada'] = 'Falta'
        case (True, False):
            print('Saiu mas não voltou')
            df.at[row, 'Falta_analisada'] = "Saiu mas não voltou"
        case (False, True):
            print('Inconsistente')
            df.at[row, 'Falta_analisada'] = "Inconsistente"
        case (False, False):
            print('Valido')
            df.at[row, 'Falta_analisada'] = "Valido"

print(df['Falta_analisada'])
```

```python
df.to_excel('testes.xlsx', index=False)
```
